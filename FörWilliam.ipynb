```python
R0 = 2*np.array([7119, 17662])
RI = 2*np.array([[42871/(2*math.pi), 7115],[6323, 8125],[5920, 8533],[5503, 8736],[4958, 9084],[4398, 9395],[3800, 9638],[3141, 9812]])
vR = np.zeros(len(RI)+1)
for i in range(1,len(RI)+1):
    vR[i] = 1/(2*d)*(+RI[i-1,0]**2-RI[i-1,1]**2)/(R0[0]**2-R0[1]**2)
vR[0] = 0
vR = vR#/2
I = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
B = I*0.1#*3/2
```

```python
B = np.delete(B, 0)
vR = np.delete(vR, 0)
vB = np.delete(vB, 0)
coeffs = np.polyfit(B, vB, 1)
print(coeffs)
fit_line = np.polyval(coeffs, B)

plt.plot(B, fit_line, 'b')
plt.plot(B, vB, 'o',color = 'b',  label = '470 nm')

plt.plot(B, vR, 'o',color = 'r',  label = '647 nm')
coeffsR = np.polyfit(B, vR, 1)
print(coeffsR)
fit_line = np.polyval(coeffsR, B)

plt.plot(B, fit_line, 'r')
B = np.insert(B, 0, 0)
myB = 9.274*10**-24 #* math.pi
h = 6.626*10**-34
c = 3*10**8
plt.plot(B, myB/(h*c)*B*2*2, 'c', label = 'Teoretisk linje för 470 nm')
plt.plot(B, myB/(h*c)*B*2, 'm', label = 'Teoretisk linje för 647 nm')


plt.legend()
plt.xlabel('Fältstyrka [T]')
plt.ylabel('Förändring i vågtal [m^-1]')
print(coeffs[0]/coeffsR[0])
```
